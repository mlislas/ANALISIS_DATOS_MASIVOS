### 5. Visualización de datos (Gráficos)

```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga el dataset
netflix_data = pd.read_csv("netflix_titles.csv")
```

### 5.1	Histograma: Distribución de los años de lanzamiento

Este histograma muestra cuántos títulos se lanzaron por año.

*	La mayoría del contenido corresponde a años recientes, especialmente entre 2010 y 2021.
*	Se observan muy pocos títulos previos a los años 2000.
*	Esto indica que el catálogo de Netflix está fuertemente orientado a material contemporáneo.



```python
# Histograma para la variable 'release_year'
plt.hist(netflix_data['release_year'], bins=30, color='red', edgecolor='black')
plt.title('Distribución de los años de lanzamiento')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Frecuencia')
plt.show()
```

### 5.2	Gráfico de barras: Cantidad de películas y series

Este gráfico compara el número total de películas y series.

*	Las películas representan la mayor parte del catálogo.
*	Las series son significativamente menos.
*	Esto confirma que Netflix ofrece más contenido cinematográfico que episodios televisivos.



```python
# Gráfico de barras para la variable 'type'
sns.countplot(data=netflix_data, x='type')
plt.title('Cantidad de películas y series')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.show()
```

### 5.3 Gráfico circular: Porcentaje de películas vs. series

*	Las películas ocupan la mayor proporción del catálogo (alrededor del 70%).
*	Las series representan el porcentaje restante, confirmando la tendencia vista en el gráfico de barras.



```python
# Gráfico circular para la variable 'type'
plt.pie(netflix_data["type"].value_counts(), 
        labels=netflix_data["type"].value_counts().index, 
        autopct="%1.1f%%")
plt.title("Porcentaje de películas y series")
plt.show()
```

### 5.4 Gráfico de densidad: Distribución de los años de lanzamiento. 

Este gráfico suaviza la distribución para mostrar tendencias.

*	Existe una alta concentración de títulos recientes, con un pico entre 2015 y 2020.
*	La densidad disminuye abruptamente para años anteriores, lo que indica que Netflix se enfoca en contenido moderno.
*	También muestra que prácticamente no hay contenido previo a 1950.



```python
# Gráfico de densidad para la variable 'release_year'
sns.kdeplot(data=netflix_data, x='release_year', fill=True, color='red')
plt.title('Densidad de los años de lanzamiento')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Densidad')
plt.show()
```

### 5.4 Gráfico combinado (líneas + barras): Análisis de géneros más frecuentes

El gráfico muestra que Netflix concentra la mayor parte de su catálogo en géneros universales (columna `listed_in`) como International Movies, Dramas y Comedies. Los géneros internacionales dominan en cantidad, reflejando una estrategia de distribución global. A medida que avanzan los géneros de la lista, se observa una disminución progresiva, lo que indica que categorías como Documentaries, Independent Movies, Romantic Movies y Thrillers, aunque presentes, representan una menor proporción del catálogo.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Gráfico línea-barras para Top 10 géneros más frecuentes (Gráfico combinado: Barras + Línea)

genres = dataset['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10)

plt.figure(figsize=(10,6))

# --- Barras ---
plt.bar(genres.index, genres.values, alpha=0.6, label='Cantidad')

# --- Línea sobre las barras ---
plt.plot(genres.index, genres.values, marker='o', linewidth=2, color='black', label='Tendencia')

plt.title('Top 10 géneros más frecuentes')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### 5.5 Gráfico con barras horizontales: Top 15 países con más contenido en Netflix

El análisis de la columna `country` revela una clara disparidad en la distribución geográfica de los títulos disponibles.

Los países con la mayor cantidad de contenido en el catálogo son:

1.  **Estados Unidos** (Dominante, con una amplia diferencia sobre el resto).
2.  **India**
3.  **Reino Unido**
4.  **Canadá**
5.  **Francia**
6.  **Japón**
7.  **España**
8.  **Corea del Sur**
9.  **México**
10. **China**
11. **Australia**
12. **Alemania**
13. **Egipto**
14. **Brasil**
15. **Turquía**

Este hallazgo revela que el catálogo de Netflix está fuertemente dominado por **producciones estadounidenses**. 
El resto del Top 5 está compuesto por industrias de cine y TV muy activas a nivel global, como **India** (Bollywood), **Reino Unido**, **Canadá** y **Francia**. 




```python
# Gráfico de barras horizontales: Países con mayor cantidad de producciones en Netflix 

countries = dataset['country'].dropna().str.split(', ', expand=True).stack().value_counts().head(15)

plt.figure(figsize=(10,6))
countries.plot(kind='barh')
plt.title('Top 15 países con más contenido en Netflix')
plt.xlabel('Cantidad')
plt.ylabel('País')
plt.gca().invert_yaxis()  # Para que el país con más títulos aparezca arriba
plt.tight_layout()
plt.show()
```

* Géneros universales e internacionales: Los géneros más frecuentes son International Movies, Dramas y Comedies. El dominio de los géneros internacionales refleja una estrategia de distribución global.
* Concentración Geográfica: El catálogo está fuertemente dominado por producciones de Estados Unidos, seguido por industrias cinematográficas y de TV activas a nivel global como India (Bollywood), Reino Unido, Canadá y Francia.
